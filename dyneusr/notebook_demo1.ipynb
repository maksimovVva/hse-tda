```python
from dyneusr import DyNeuGraph
from dyneusr.datasets import make_trefoil
from dyneusr.tools import visualize_mapper_stages
from kmapper import KeplerMapper
from kmapper.cover import Cover
import sklearn
import pandas as pd
```

```python
from sklearn import datasets
digits = datasets.load_digits()
```

```python
X = digits.data
```

```python
target = digits.target
labels = digits.target_names
```

```python
y = pd.DataFrame({l:(target==i).astype(int) for i,l in enumerate(labels)})
```

```python
mapper = KeplerMapper(verbose=1)
lens = mapper.fit_transform(X, projection=sklearn.manifold.TSNE())
graph = mapper.map(lens, clusterer=sklearn.cluster.DBSCAN(metric="cosine"), cover=Cover(30, 0.1))
```

```python
# Visualize the shape graph using DyNeuSR's DyNeuGraph                          
dG = DyNeuGraph(G=graph, y=y)
dG.visualize('dyneusr_output.html')
```

```python
from dyneusr.tools import visualize_mapper_stages
```

```python
import matplotlib.pyplot as plt
```

```python
# Visualize the stages of Mapper
fig, axes = visualize_mapper_stages(
    digits, lens=lens, graph=graph, cover=mapper.cover, figsize=(16, 4))

plt.show()
```

```python
# Define projections to compare
projections = ([0], [0,1], [1,2], [0, 2])

# Compare different sets of columns as lenses
for projection in projections:

    # Generate shape graph using KeplerMapper
    mapper = KeplerMapper(verbose=1)
    lens = mapper.fit_transform(X, projection=projection)
    graph = mapper.map(lens, X, nr_cubes=4, overlap_perc=0.3)

    # Visualize the stages of Mapper
    fig, axes = visualize_mapper_stages(
        dataset, lens=lens, graph=graph, cover=mapper.cover, 
        layout="spectral", figsize=(16, 4))

    # Save each figure
    plt.savefig(
        "mapper_lens_{}.png".format("_".join(str(_) for _ in projection)),
            dpi=600, background='transparent')

# Show all figures
plt.show()
```

```python
# Visualize the shape graph using DyNeuSR's DyNeuGraph                          
dG = DyNeuGraph(G=graph, y=y)
dG.visualize('dyneusr_output.html')
```

```python
from nilearn.datasets import fetch_haxby
from nilearn.input_data import NiftiMasker
from kmapper import KeplerMapper, Cover
from sklearn.manifold import TSNE
from sklearn.cluster import DBSCAN

from dyneusr import DyNeuGraph
from dyneusr.tools import visualize_mapper_stages
```

```python
dataset = fetch_haxby()
masker = NiftiMasker(
    dataset.mask_vt[0], 
    standardize=True, detrend=True, smoothing_fwhm=4.0,
    low_pass=0.09, high_pass=0.008, t_r=2.5,
    memory="nilearn_cache"
    )
X = masker.fit_transform(dataset.func[0])
```

```python
# Encode labels as integers
df = pd.read_csv(dataset.session_target[0], sep=" ")
target, labels = pd.factorize(df.labels.values)
y = pd.DataFrame({l:(target==i).astype(int) for i,l in enumerate(labels)})
```

```python
mapper = KeplerMapper(verbose=1)
lens = mapper.fit_transform(X, projection=TSNE(2, random_state=1))
graph = mapper.map(
    lens, X=X, 
    cover=Cover(20, 0.5), 
    clusterer=DBSCAN(eps=20.)
    )
```

```python
# Visualize the shape graph using DyNeuSR's DyNeuGraph
dG = DyNeuGraph(G=graph, y=y)
dG.visualize('dyneusr_haxby_decoding.html', static=True, show=True)
```

```python
fig, axes = visualize_mapper_stages(
    dataset, y=target, lens=lens, graph=graph, cover=mapper.cover, 
    node_size=20, edge_size=0.5, edge_color='gray',
    layout="kamada_kawai",  figsize=(16, 4)
    )
plt.show()
```
